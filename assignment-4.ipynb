```python
class TestAssignmentFour(unittest.TestCase):
    def test_01_Hogwarts(self):
        hogwarts = Hogwarts()
        self.assertIsInstance(hogwarts, Hogwarts)
        self.assertIsInstance(hogwarts.location, str)
        self.assertIsInstance(hogwarts.founders, list)
        self.assertTrue(len(hogwarts.founders) == 4)
        self.assertIsInstance(hogwarts.houses, list)
        self.assertTrue(len(hogwarts.houses) == 4)
    def test_02_Gryffindor(self):
        harry_potter = Gryffindor("Harry Potter")
        self.assertIsInstance(harry_potter, Gryffindor)
        self.assertIsInstance(harry_potter.name, str)
        self.assertIsInstance(harry_potter.traits, list)
        self.assertIsInstance(harry_potter.colors, list)
        self.assertEqual(harry_potter.cast_a_spell(), 'Expelliarmus!')
        hermione_granger = Gryffindor("Hermione Granger")
        self.assertIsInstance(hermione_granger, Gryffindor)
        self.assertIsInstance(hermione_granger.name, str)
        self.assertIsInstance(hermione_granger.traits, list)
        self.assertIsInstance(hermione_granger.colors, list)
        self.assertEqual(hermione_granger.cast_a_spell(), 'Expelliarmus!')
    def test_03_Palindrome(self):
        palindrome = Palindrome('eye')
        self.assertIsInstance(palindrome, Palindrome)
        self.assertEqual(palindrome.original_text, 'eye')
        self.assertEqual(palindrome.reversed_text, 'eye')
        self.assertTrue(palindrome.is_palindrome())
        palindrome = Palindrome('dye')
        self.assertIsInstance(palindrome, Palindrome)
        self.assertEqual(palindrome.original_text, 'dye')
        self.assertEqual(palindrome.reversed_text, 'eyd')
        self.assertFalse(palindrome.is_palindrome())
        palindrome = Palindrome('did')
        self.assertIsInstance(palindrome, Palindrome)
        self.assertEqual(palindrome.original_text, 'did')
        self.assertEqual(palindrome.reversed_text, 'did')
        self.assertTrue(palindrome.is_palindrome())
    def test_04_MinMaxFinder(self):
        min_max_finder = MinMaxFinder([2, 3, 5, 7, 11, 11, 7, 5, 3, 2])
        self.assertIsInstance(min_max_finder, MinMaxFinder)
        self.assertIsInstance(min_max_finder.integer_list, list)
        self.assertEqual(min_max_finder.find_min(), 2)
        self.assertEqual(min_max_finder.find_max(), 11)
        self.assertEqual(min_max_finder.find_idxmin(), [0, 9])
        self.assertEqual(min_max_finder.find_idxmax(), [4, 5])
        min_max_finder = MinMaxFinder([10, 9, 8, 6, 4, 1, 1, 4, 6, 8, 9, 10])
        self.assertIsInstance(min_max_finder, MinMaxFinder)
        self.assertIsInstance(min_max_finder.integer_list, list)
        self.assertEqual(min_max_finder.find_min(), 1)
        self.assertEqual(min_max_finder.find_max(), 10)
        self.assertEqual(min_max_finder.find_idxmin(), [5, 6])
        self.assertEqual(min_max_finder.find_idxmax(), [0, 11])
    def test_05_CommonDivisors(self):
        cd = CommonDivisors(3, 6)
        self.assertIsInstance(cd, CommonDivisors)
        self.assertEqual(len(cd.x_divisors), 2)
        self.assertEqual(len(cd.y_divisors), 4)
        self.assertEqual(len(cd.get_common_divisors()), 2)
        cd = CommonDivisors(4, 8)
        self.assertIsInstance(cd, CommonDivisors)
        self.assertEqual(len(cd.x_divisors), 3)
        self.assertEqual(len(cd.y_divisors), 4)
        self.assertEqual(len(cd.get_common_divisors()), 3)
    def test_06_PrimeJudger(self):
        pj = PrimeJudger(1)
        self.assertIsInstance(pj, PrimeJudger)
        self.assertEqual(len(pj.get_divisors()), 1)
        self.assertFalse(pj.is_prime())
        pj = PrimeJudger(2)
        self.assertIsInstance(pj, PrimeJudger)
        self.assertEqual(len(pj.get_divisors()), 2)
        self.assertTrue(pj.is_prime())
        pj = PrimeJudger(4)
        self.assertIsInstance(pj, PrimeJudger)
        self.assertEqual(len(pj.get_divisors()), 3)
        self.assertFalse(pj.is_prime())
    def test_07_ArrayGetAttrs(self):
        arr = np.array([5, 5, 6, 6])
        aga = ArrayGetAttrs(arr)
        self.assertIsInstance(aga, ArrayGetAttrs)
        self.assertEqual(aga.get_ndim(), 1)
        self.assertEqual(aga.get_shape(), (4,))
        self.assertEqual(aga.get_size(), 4)
        arr = np.array([[5, 5],
                        [6, 6]])
        aga = ArrayGetAttrs(arr)
        self.assertIsInstance(aga, ArrayGetAttrs)
        self.assertEqual(aga.get_ndim(), 2)
        self.assertEqual(aga.get_shape(), (2, 2))
        self.assertEqual(aga.get_size(), 4)
    def test_08_add_intercepts(self):
        A = np.array([5, 5, 6, 6]).reshape(-1, 1)
        np.testing.assert_array_equal(add_intercepts(A),
        np.array([[1, 5],
                  [1, 5],
                  [1, 6],
                  [1, 6]]))
        B = np.arange(10).reshape(5, 2)
        np.testing.assert_array_equal(add_intercepts(B),
        np.array([[1, 0, 1],
                  [1, 2, 3],
                  [1, 4, 5],
                  [1, 6, 7],
                  [1, 8, 9]]))
        C = np.array([5, 6, 7, 8]).reshape(-1, 1)
        np.testing.assert_array_equal(add_intercepts(C),
        np.array([[1, 5],
                  [1, 6],
                  [1, 7],
                  [1, 8]]))
    def test_10_create_diagonal_split_matrix(self):
        np.testing.assert_array_equal(create_diagonal_split_matrix(2, 5566),
        np.array([[   0, 5566],
                  [5566,    0]]))
        np.testing.assert_array_equal(create_diagonal_split_matrix(3, 55),
        np.array([[ 0, 55, 55],
                  [55,  0, 55],
                  [55, 55,  0]]))
        np.testing.assert_array_equal(create_diagonal_split_matrix(4, 66),
        np.array([[ 0, 66, 66, 66],
                  [66,  0, 66, 66],
                  [66, 66,  0, 66],
                  [66, 66, 66,  0]]))
    def test_09_create_square_matrix(self):
        np.testing.assert_array_equal(create_square_matrix(3),
        np.array([[1, 2, 3],
                  [2, 4, 6],
                  [3, 6, 9]]))
        np.testing.assert_array_equal(create_square_matrix(4),
        np.array([[ 1,  2,  3,  4],
                  [ 2,  4,  6,  8],
                  [ 3,  6,  9, 12],
                  [ 4,  8, 12, 16]]))
        np.testing.assert_array_equal(create_square_matrix(5),
        np.array([[ 1,  2,  3,  4,  5],
                  [ 2,  4,  6,  8, 10],
                  [ 3,  6,  9, 12, 15],
                  [ 4,  8, 12, 16, 20],
                  [ 5, 10, 15, 20, 25]]))

suite = unittest.TestLoader().loadTestsFromTestCase(TestAssignmentFour)
runner = unittest.TextTestRunner(verbosity=2)
test_results = runner.run(suite)
number_of_failures = len(test_results.failures)
number_of_errors = len(test_results.errors)
number_of_test_runs = test_results.testsRun
number_of_successes = number_of_test_runs - (number_of_failures + number_of_errors)
```
